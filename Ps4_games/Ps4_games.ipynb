```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
df = pd.read_csv("List of best-selling PlayStation 4 video games.csv")
df
```

```python
df.info()
```

```python
df.rename(columns= {"Copies sold":"Copies_sold(millions)",
                    "Release date[a]": "Release_date"}, 
          inplace= True)
```

```python
df["Game"] = df["Game"].astype(str)
```

```python
df["Copies_sold(millions)"] = df["Copies_sold(millions)"] .str.replace('million', '')
```

```python
df["Copies_sold(millions)"] = df["Copies_sold(millions)"].astype(float)
```

```python
df["Release_date"] = df["Release_date"].astype("datetime64[ns]")
```

```python
df["Genre(s)"] = df["Genre(s)"].astype(str)
```

```python
df["Developer(s)"] = df["Developer(s)"].astype(str)
```

```python
df["Publisher(s)"] = df["Publisher(s)"].astype(str)
```

```python
df.info()
```

```python
df.sort_values(by= "Copies_sold(millions)", ascending= False)
```

```python
Dev_games = df.groupby('Developer(s)')['Game'].count().sort_values(ascending=False)

plt.figure(figsize= (12,6))
sns.barplot( x = Dev_games.values,
            y = Dev_games.index,
            palette="Greys")
sns.set(rc={'axes.facecolor':'Grey',
            'figure.facecolor':'Lightgrey'})
plt.xlabel('Count', weight = 'bold')
plt.ylabel('Developer(s)', weight ='bold')
plt.yticks(weight = 'bold')
plt.xticks(weight = 'bold')
plt.show()
```

```python
plt.figure(figsize= (12,6))
sns.barplot( x =df['Copies_sold(millions)'],
            y = df['Developer(s)'],
            palette="Greys")
sns.set(rc={'axes.facecolor':'Grey',
            'figure.facecolor':'Lightgrey'})
plt.xlabel('Copies sold in millions', weight = 'bold')
plt.ylabel('Genre(s)',weight ='bold')
plt.yticks(weight = 'bold')
plt.xticks(weight = 'bold')
plt.show()
```
