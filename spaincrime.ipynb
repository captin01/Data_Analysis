- Importing relevant libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

- Visual representation of crimes per year

```python
plt.figure(figsize = (9,6))
plt.bar( x = s_data['Year'],
        height = s_data['Total cases'],
        color = "grey")
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.ylabel("No.of totalcases" , fontsize = 10)
plt.xlabel("Years", fontsize = 12)
plt.title("Crime_rate per year", fontsize = 12, fontweight = "bold")
plt.show()
```
